```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-2, 2, 100)

# Compute 2^x for each x value
y = 2 ** x

# Create the plot
plt.plot(x, y, label='$2^x$')

# Add title and labels
plt.title('Graph of $2^x$')
plt.xlabel('x')
plt.ylabel('$2^x$')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
